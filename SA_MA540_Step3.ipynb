**The First Quartile:** The first quartile is the point that partitions the first 25% of the data and later 75%.
```python
q2=np.quantile(data,0.50)
print('The second quantile of the data is:',round(q2,4))
```

**The Third Quartile:** The third quartile is the point that partitions the data in 75-25 ratio. ## CRV-Central Limit Theorem Verification
 
Verify the Central Limit Theorem by taking random samples from the simulated data and calculating sample means. You can observe how
the sample means approximate a normal distribution and understand the importance of the
CLT in statistics.

```python
# define a function to obtain the sample means for given population
# pass the population and and number of samples as the input parameters
def sample_mean_calculator(data, n_samples):
    
    # create an empty list to store the means of each sample
    sample_means = []
    for i in range(n_samples):
       
 # generates a random sample of size 500 from a given array or a list
        sample = np.random.choice(data, size=500, replace=False)
        
        # calculate the mean of the sample data
        sample_mean = np.mean(sample)
        
        # append the mean to the list
        sample_means.append(sample_mean)    
        
    return sample_means
```

<a id="zs"> </a>
####  CRV-Z-Score

Another way to detect outliers are using the z-score
<ul>
  <li>Z-score of a value is the difference between that value and the mean, divided by the standard deviation</li>
  <li>Z-score of 0 indicates the value is the same as the mean</li>
  <li>Z-score is positive or negative means the value is above or below the mean and by how many standard deviations</li>
  <li>Z-score greater than 3 or less than -3, indicates an outlier value</li>
</ul>

```python
upper_limit = mean + 2.73 * std
lower_limit = mean - 2.73 * std
data[((data > upper_limit) | (data < lower_limit))]
print(outliers)
```

```python
# Z-scores are defined for each observation in a variable
# compute the z-scores using the method z-score from the scipy library 
zscores_data = scipy.stats.zscore(data)

# print the z-score
zscores_data[:100]
```

**Interpretation:** The above output gives the z-scores which are less than -3 and greater than 3. The values corresponding to these score are considered to be outliers

## CRV-Probability Calculations
Calculate probabilities related to the continuous (such as normal) distribution. For example, you can find the probability that a randomly selected value
falls within a specified range or above/below a certain threshold.

Let us find P(-0.2 <= X <= 0.1)

```python
# standardize the variable with x = -0.2
z_2 = ((-.2 - mean) / std)
# 'cdf()' returns the P(Z <= z) i.e P(Z <= -0.2)
p_2= stats.norm.cdf(z_2)
# standardize the variable with x = 0.1
z_1 = (0.1 - mean) / std

# 'cdf()' returns the P(Z <= z) i.e P(Z <= 0.1)
p_1 = stats.norm.cdf(z_1)

# calculate the difference between 'p_1' and 'p_2' to find the required probability
prob = p_1 - p_2

# use 'round()' to round-off the value to 4 digits
req_prob = prob
```

```python
print('The probability is', round(req_prob,4))
```

```python
transition_matrix = np.array([
[0.7, 0.2, 0.1],
[0.3, 0.6, 0.1],
[0.2, 0.3, 0.5]
])
# Define the initial state probabilities
initial_state_probabilities = np.array([0.3, 0.4, 0.3])
# Number of steps in the simulation
num_steps = 10
# Simulate the Markov chain
current_state = np.random.choice(range(len(initial_state_probabilities)),p=initial_state_probabilities)
for step in range(num_steps):
    print(f"Step {step + 1}: State {current_state + 1}")
    current_state = np.random.choice(range(len(transition_matrix)),p=transition_matrix[current_state])
# Calculate state probabilities after the specified number of steps
# initial_state_prob (I) * transition_matrix (T)^(num_steps[10])
# I.T^10
final_state_probabilities = np.dot(initial_state_probabilities, np.linalg.matrix_power(transition_matrix, num_steps))
# Print the final state probabilities
print(f"\nState Probabilities after {num_steps} steps:")
for i, prob in enumerate(final_state_probabilities):
    print(f"State {i + 1}: {prob:.4f}")
```

We simulate the Markov chain by randomly selecting the next state based on the transition
probabilities.

```python
import numpy as np

class RecurrentEvent:
    def __init__(self, transition_matrix):
        self.transition_matrix = transition_matrix
        self.number_states = len(transition_matrix)
        self.state = 0

    def event(self):
        next_state = np.random.choice(range(self.number_states), p=self.transition_matrix[self.state])
        self.state = next_state

# Example Usage
# Define the transition matrix
# Rows represent current states, and columns represent next states
transition_matrix = np.array([[0.0, 0.2, 0.8],
                              [0.1, 0.6, 0.3],
                              [0.2, 0.3, 0.5]])

recurrentmodel = RecurrentEvent(transition_matrix)

# Simulate multiple events
num_simulations = 15
for _ in range(num_simulations):
    recurrentmodel.event()
    print("Current State:", recurrentmodel.state)
```

Recurrent events in a Markov chain refer to the situation where the system repeatedly
returns to certain states.Let's consider a simple example where we model the states of a
system with three recurrent events: event 0,event 1 and event 2. The Markov chain can
transition between these events based on specified transition probabilities.


```python
#Ergodicity
import numpy as np

transition_matrix = np.array([[0.5, 0.2, 0.3],
                              [0.1, 0.6, 0.3],
                              [0.2, 0.3, 0.5]])

num_states = 3

# Initial state
current_state = np.random.choice(range(num_states))

num_steps = 1000
sequences = []
state_probabilities = np.zeros(num_states)

# Simulate Markov chain
for i in range(num_steps):
    sequences.append(current_state)
    current_state = np.random.choice(range(num_states), p=transition_matrix[current_state])

for sequence in sequences:
    state_probabilities[sequence] += 1

state_probabilities /= num_steps  # This is the time-averaged probability

# Calculate stationary distribution (also known as equilibrium distribution)
stationary_distribution = np.linalg.matrix_power(transition_matrix, 1000)[0]

# Display results
print("Empirical Probabilities:", state_probabilities)
print("Stationary Distribution:", stationary_distribution)
```

In the above example we see by comparing the time-averaged behavior (empirical
probabilities) with the stationary distribution of the Markov chain. In an ergodic Markov
chain, the time-averaged behavior should converge to the stationary distribution

## Variance Reduction Techniques

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to estimate
def target_function(x):
    return x**2

# Analytical solution (for comparison)
def analytical_solution():
    return 1/3

# Monte Carlo simulation without variance reduction
def monte_carlo_no_variance_reduction(num_samples):
    samples = np.random.uniform(0, 1, num_samples)
    estimate = np.mean(target_function(samples))
    return estimate

# Control variates Monte Carlo simulation
def monte_carlo_with_control_variates(num_samples, control_variate_factor=0.5):
    samples = np.random.uniform(0, 1, num_samples)
    control_variate = samples**2  # Potential control variate (known in this case)
    
    target_values = target_function(samples)
    control_variate_values = control_variate_function(control_variate)

    estimate = np.mean(target_values - control_variate_factor * (control_variate_values - analytical_solution()))
    return estimate

# Control variate function (known in this case)
def control_variate_function(x):
    return x**2

# Number of Monte Carlo samples
num_samples = 10000

# Perform Monte Carlo simulations
estimate_no_variance_reduction = monte_carlo_no_variance_reduction(num_samples)
estimate_with_control_variates = monte_carlo_with_control_variates(num_samples)

# Print results
print("Monte Carlo Estimate (No Variance Reduction):", estimate_no_variance_reduction)
print("Monte Carlo Estimate (With Control Variates):", estimate_with_control_variates)
print("Analytical Solution:", analytical_solution())

# Plotting
plt.bar(["No Variance Reduction", "With Control Variates", "Analytical Solution"],
        [estimate_no_variance_reduction, estimate_with_control_variates, analytical_solution()])
plt.ylabel("Estimate")
plt.title("Monte Carlo Estimation with Control Variates")
plt.show()
```

## Combinatorial Analysis

```python
import itertools
import random

# Function to simulate a combinatorial problem (random permutation)
def simulate_combinatorial_problem(elements):
    # Generate all permutations
    all_permutations = list(itertools.permutations(elements))
    
    # Randomly select a permutation
    random_permutation = random.choice(all_permutations)
    
    return random_permutation

# Set of elements
elements = [1, 2, 3, 4]

# Perform the simulation
random_permutation = simulate_combinatorial_problem(elements)

# Print the result
print("Random Permutation:", random_permutation)
```

In this example, itertools.permutations is used to generate all permutations of the given set of elements. The simulate_combinatorial_problem function then randomly selects one permutation from the list of all permutations using random.choice.
